## Data Dropped

We dropped data from both dataframes in order to **only** look at data that is relevant to our current objective: 

> Are employees who have only worked for the institutes for a short period of time resigning due to some kind of dissatisfaction? What about employees who have been at the job longer?

To ensure this, we are looking for instances where the `separationtype` is Resignation.

```python
dete_resignations['cease_date'].value_counts()
```

```python
pattern = '[0-9]*\/*([0-9]{4})'
dete_resignations['cease_date'] = dete_resignations['cease_date'].str.extract(pattern)
dete_resignations['cease_date'] = dete_resignations['cease_date'].astype(float)
```

We have cleaned the data to reflect only the years of resignation

DETE dataset now has a column for years of service for each employee. This is helpful to segregate people who have quit in a short span from people who have worked for longer.

```python
def values(value):
    if value == '-':
        return False
    elif pd.isnull(value):
        return np.nan
    else:
        return True
```

```python
tafe_resignations['dissatisfied'] = tafe_resignations[['Contributing Factors. Dissatisfaction', 'Contributing Factors. Job Dissatisfaction']].applymap(values).any(axis = 1, skipna=False)
tafe_resignations_up = tafe_resignations.copy()
```

```python
tafe_resignations['dissatisfied'].value_counts()
```

```python
tafe_resignations_up = tafe_resignations.copy()
```

```python
dete_resignations['dissatisfied'] = dete_resignations['dissatisfied'] = dete_resignations[['job_dissatisfaction',
       'dissatisfaction_with_the_department', 'physical_work_environment',
       'lack_of_recognition', 'lack_of_job_security', 'work_location',
       'employment_conditions', 'work_life_balance',
       'workload']].any(1, skipna=False)
```

```python
dete_resignations_up = dete_resignations.copy()
```

```python
dete_resignations_up['dissatisfied'].value_counts()
```

```python
dete_resignations_up['institute'] = 'DETE'
tafe_resignations_up['institute'] = 'TAFE'
```

```python
combined = pd.concat([dete_resignations_up, tafe_resignations_up], ignore_index=True)
```

```python
combined_updated = combined.dropna(axis = 1, thresh = 500).copy()
```

```python
combined['institute_service'].value_counts()
```

```python
combined_updated['institute_service'] = combined_updated['institute_service'].astype('str')
```

```python
pattern = '([\d]+)'
combined_updated['institute_service'] = combined_updated['institute_service'].str.extract(pattern)
combined_updated['institute_service'] = combined_updated['institute_service'].astype('float')
```

```python
combined_updated['institute_service'].value_counts()
```

```python
def category(element):
    if element >= 11:
        return 'Veteran'
    elif element >= 7:
        return 'Established'
    elif element >= 3:
        return 'Experienced'
    elif pd.isnull(element):
        return np.nan
    else:
        return 'New'
    
combined_updated['service_cat'] = combined_updated['institute_service'].apply(category)
```
