## Clean-Up Required

1. Price column has the dollar symbol and is a string
2. Odomoeter column has "km" at the end.
3. Offer type is "Angebot" for all except one entry. This column can be dropped.
4. Dates are stored as Text or Numeric values which needs to be modified.
5. No. of pictures is zero
6. Seller is almost always privat with one exception

```python
autos['price'] = autos['price'].str.replace("$","").str.replace(',','').astype(int)
```

```python
autos['odometer'] = autos['odometer'].str.replace('km','').str.replace(',','').astype(int)
```

```python
autos.rename({'odometer':'odometer_km'},axis = 1, inplace = True)
```

```python
autos = autos.drop(['no_of_pictures','seller','offer_type'],axis = 1)
```

## Registration Year

Looking at the data, we can consider values from  1900 to 2019.

```python
autos['registration_year'].value_counts().sort_index().head(15)
```

## Observations for Registration Year

Most cars were registered relatively recently. ~ 10-15 years

```python
autos['brand'].value_counts(normalize = True).sort_index(ascending = True)
```

```python
brand_counts = autos['brand'].value_counts(normalize = True)
common_brands = brand_counts[brand_counts > 0.05].index
```

```python
common_brands
```

```python
mean_prices = {}

for brand in common_brands:
    mean = autos.loc[autos['brand'] == brand,'price'].mean()
    mean_prices[brand] = mean
    
```

```python
mean_prices
```

```python
mean_mileage ={}
for brand in common_brands:
    mean = autos.loc[autos['brand'] == brand,'odometer_km'].mean()
    mean_mileage[brand] = mean
```

```python
pd.Series(mean_prices)
```

```python
pd.Series(mean_mileage)
```

```python
df = pd.DataFrame(pd.Series(mean_prices))
```

```python
df['odometer'] = pd.Series(mean_mileage)
```
